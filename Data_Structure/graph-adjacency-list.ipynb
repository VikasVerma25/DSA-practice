```python
class LinkedIn:
    def __init__(self):
        self.graph = dict()
        
    def addprofile(self, profile):
        self.graph[profile] = []
        
    def addconnection(self, profile, connection):
        if connection not in self.graph[profile]:
            self.graph[profile].append(connection)
        else:
            return "Already connected"
    
    def showconnections(self, profile):
        return self.graph[profile]
    
    def allprofiles(self):
        return self.graph.keys()
```

```python
mygraph = LinkedIn()
```

```python
mygraph.addprofile('jack')
mygraph.addprofile('eric')
mygraph.addprofile('james')
mygraph.addprofile('tom')
mygraph.addprofile('john')
```

```python
mygraph.addconnection('john', 'james')
```

```python
mygraph.showconnections('jack')
```

```python
mygraph.allprofiles()
```
