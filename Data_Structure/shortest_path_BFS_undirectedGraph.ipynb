```python
class graph:
    def __init__(self, gdict):
        self.gdict = gdict
    
    def spp_bfs(self, start, end):
        queue = []
        queue.append([start])
        while True:
            path = queue.pop(0)
            node = path[-1]
            print("path:",path,"node:",node)
            if node == end:
                return path
            for adjacent in self.gdict.get(node):
                newpath = path.copy()
                newpath.append(adjacent)
                queue.append(newpath)
                print(queue)
```

```python
dag = graph({
    'A':['B','C'],
    'B':['D','G'],
    'C':['D','E'],
    'D':['F'],
    'E':['F'],
    'G':['F'],
})
```

```python
dag.spp_bfs('A','E')
```
