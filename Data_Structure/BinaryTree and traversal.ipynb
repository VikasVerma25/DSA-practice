```python
#Pre-order traversing using Recursion Root-Left-Right
```

```python
def preOrderTraversal(mytree):
    if not mytree:
        return
    print(mytree.data, end=" ")
    preOrderTraversal(mytree.leftchild)
    return preOrderTraversal(mytree.rightchild)
```

```python
preOrderTraversal(root)
```

```python
#Post-order traversal using Recursion   Left-Right-Root
def postOrderTraversal(mytree):
    if not mytree:
        return
    postOrderTraversal(mytree.leftchild)
    postOrderTraversal(mytree.rightchild)
    print(mytree.data, end=" ")
    
```

```python
postOrderTraversal(root)
```

```python
#In-order traversal using Recursion   Left-Root-Right
def inOrderTraversal(mytree):
    if not mytree:
        return
    inOrderTraversal(mytree.leftchild)
    print(mytree.data, end=" ")
    inOrderTraversal(mytree.rightchild)
```

```python
inOrderTraversal(root)
```

```python
#Level order traversing of tree using queue
from collections import deque
class Node:
    def __init__(self, data):
        self.data = data
        self.leftchild = None
        self.rightchild = None
    
    @staticmethod
    def level_order_traversal(rootnode):
        if rootnode == None:
            return "No node Found"
        q = deque()
        q.append(rootnode)
        
        while len(q) != 0:
            p = q.popleft()
            print(p.data)
            if p.leftchild != None:
                q.append(p.leftchild)
            if p.rightchild != None:
                q.append(p.rightchild)
        
```

```python
A = Node('A')
B = Node('B')
C = Node('C')
D = Node('D')
E = Node('E')
```

```python
A.leftchild = B
A.rightchild = C
B.leftchild = D
B.rightchild = E
```

```python
Node.level_order_traversal(A)
```
