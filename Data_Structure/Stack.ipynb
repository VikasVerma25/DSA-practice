```python
# Stack Class - object as stack - using LinkedList
class Node:
    def __init__(self, value):
        self.data = value
        self.next = None
class Stack:
    def __init__(self):
        self.top = None
        self.bottom = None
        self.length = 0
        
    def push(self, value):     # enqueue
        node = Node(value)
        if self.top == None:
            self.bottom = self.top = node
        else:            
            node.next = self.top
            self.top = node
        self.length += 1
            
    def pop(self):                  # dequeue
        if self.length == 1:
            self.bottom = self.top = None
        else:
            temp = self.top.next
            self.top.next = None
            self.top = temp
        self.length -= 1
    
    def display(self):
        temp = self.top
        if temp == None:
            return "Empty"
        while temp:
            print(temp.data)
            temp = temp.next
            
```

```python
s = Stack()
s.push(10)
s.push(201)
s.push(50)
s.display()
```

```python
# stack using array
class stack:
    def __init__(self):
        self.arr = []
        
    def peek(self):
        if len(self.arr)>0:
            return self.arr[-1]
        return "empty stack"
    
    def pop(self):
        if len(self.arr)>0:
            return self.arr.pop()
        return "empty stack"
    
    def push(self, value):
        self.arr.append(value)
```

```python
mystack = stack()
```

```python
mystack.push(30)
```
