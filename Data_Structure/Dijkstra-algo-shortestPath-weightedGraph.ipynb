```python
from collections import defaultdict
```

```python
class graph:
    def __init__(self):
        self.edges = defaultdict(list)
        self.weights = dict()
        self.nodes = set()
    def addNode(self,node):
        self.nodes.add(node)
    def addEdge(self, fromnode, tonode, weight):
        self.weights[(fromnode,tonode)] = weight
        self.edges[fromnode].append(tonode)
        
```

```python
'''
       B
   5 /   \3
    /     \
  A/_6_C_1_\E
   \       /
  4 \     /7
     \ D /
'''
g = graph()
g.addNode('A')
g.addNode('B')
g.addNode('C')
g.addNode('D')
g.addNode('E')
g.addEdge('A','B',5)
g.addEdge('A','C',6)
g.addEdge('A','D',4)
g.addEdge('B','E',3)
g.addEdge('C','E',1)
g.addEdge('D','E',5)
```

```python
def my_dijkstra(graph, initial):
        visited = {initial:0}
        path = defaultdict(list)
        nodes = set(graph.nodes)
        #O(V^2) time, O(E) space
        while nodes:
            minNode = None
            for node in nodes:
                if node in visited:
                    if minNode is None:
                        minNode = node
                    elif visited[node] < visited[minNode]:
                        minNode = node
            if minNode is None:
                break
            nodes.remove(minNode)
            currentWeight = visited[minNode]
            for edge in graph.edges[minNode]:
                weight = currentWeight + graph.weights[(minNode,edge)]
                if edge not in visited or weight<visited[edge]:
                    visited[edge] = weight
                    path[edge].append(minNode)
        return visited,path   
```

```python
my_dijkstra(g,'A')
```
