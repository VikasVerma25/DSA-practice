```python
class MyNode:
    def __init__(self, data):
        self.data = data
        self.children = []
        
    def addChild(self, c):
        self.children.append(c)
```

```python
drinks = MyNode("drinks")     #[] empty children list
```

```python
hot = MyNode("hot")
cold = MyNode("cold")
```

```python
drinks.__dict__
```

```python
drinks.addChild(hot)
drinks.addChild(cold)
```

```python
drinks.__dict__
```

```python
cold.__dict__
```

```python
tea = MyNode("tea")
coffee = MyNode("coffee")
juice = MyNode("juice")
soda = MyNode("soda")
cola = MyNode("cola")
```

```python
hot.addChild(tea)
hot.addChild(coffee)
cold.addChild(juice)
cold.addChild(soda)
cold.addChild(cola)
```

```python
for i in hot.children:
    print(i.data)
```

```python
for i in cold.children:
    print(i.data)
```

```python
# traversing tree
print(drinks.data,":")
for i in drinks.children:
    print("\t",i.data,":")
    for j in i.children:
        print("\t  ",j.data)
```
