```python
#creating Queue using array
class Queue:
    def __init__(self):
        self.items = []
    def enqueue(self, data):
        self.items.insert(0, data)
    def size(self):
        return len(self.items)
    def isEmpty(self):
        return self.items == []
    def peek(self):
        return self.items[-1]
    def dequeue(self):
        if not self.isEmpty():
            return self.items.pop()
        return "Queue is Empty"
```

```python
#creating Queue using Linked list
class MyNode:
    first = None
    last = None
    def __init__(self, data):
        self.data = data
        self.nextnode = None
        if MyNode.first == None and MyNode.last == None:
            MyNode.first = self
            MyNode.last = self
            
        else:
            MyNode.last.nextnode = self
            MyNode.last = self
    @staticmethod
    def display():
        head = MyNode.first
        if head == None:
            return "Nothing to display"
        while True:
            print(head.data, end=" <-- ")
            head = head.nextnode      
            if head == None:
                break
    @staticmethod
    def dequeue():
        if MyNode.first == None and MyNode.last == None:
            return "Queue is empty"
        elif MyNode.first == MyNode.last:
            d = MyNode.first.data
            MyNode.first, MyNode.last = None, None
        else:
            d = MyNode.first.data
            MyNode.first = MyNode.first.nextnode
        return d
```

```python
### queue using Linked List - object as a queue
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
        
class Queue:
    def __init__(self):
        self.first = None
        self.last = None
        self.length = 0
        
    def enqueue(self, value):
        node = Node(value)
        if self.first is None:
            self.first = self.last = node   
        else:
            self.last.next = node
            self.last = node
        self.length += 1
        
    def dequeue(self):
        if self.length == 0:
            return "empty queue"
        if self.length == 1:
            temp = self.first
            self.first = self.last = None
        else:
            temp = self.first
            self.first = self.first.next
            temp.next = None
            
        self.length -= 1
        return temp.value   
    
    def display(self):
        temp =self.first
        if not temp:
            return "Empty queue"
        while temp:
            print(temp.value," <= ", end='')
            temp = temp.next
        
        
```

```python
myq = Queue()
```

```python
myq.enqueue(47)
```

```python
myq.dequeue()
```
