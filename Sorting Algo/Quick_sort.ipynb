# Divide and Conquer

An element is at right positoin if all left elments of it are less than it and all right elements are greater than it. 

```python
def quick(arr, l, r):
    if l < r:
        j = partition(arr,l,r)
        quick(arr,l,j-1)
        quick(arr,j+1,r)
        
def partition(arr,left,right):
    pivot = arr[left]
    i,j = left,right
    while i < j:
        while pivot >= arr[i] and i<j:      # i<j condition, in case it goes till end of array
            i += 1
        while pivot < arr[j]:
            j -= 1
        if i < j:
            arr[i],arr[j] = arr[j],arr[i]
    else:
        arr[left],arr[j]=arr[j],arr[left]
    return j   
```

```python
arr = [6,2,8,5,1,9,4]
quick(arr,0,len(arr)-1)
print(arr)
```
