Divide and Conquer   

```python
"""
(1.) keep dividing into two subarrays until we reach single elements. (Recursion) -> Divide
1 to 12 is sequence in which execution occurs.
            [2,7,1,5,9,4,5]
            /              \
      1.[2,7,1,5]           8.[9,4,5]
      /         \                /   \
 2.[2,7]        5.[1,5]       9.[9,4] 12.[5]
    /  \        /   \          /    \
3.[2] 4.[7]  6.[1]  7.[5]   10.[9]  11.[4]    (l not less than r, so go back)
 
(2.) then keep combining the two soretd array. -> Conquer
1-2-3-4-5-6 is the order in which subarrays are sorted 
______________________________________________

        6. [1,2,4,5,5,7,9]          ===> Sorted
            /          \
   5.[1,2,5,7]        2.[4,5,9]
      /    \             /   \
 4.[2,7]   3.[1,5]     1.[9,4]  [5]   (called the merge function)
    / \     /   \      /   \ 
[2]  [7]  [1]  [5]   [9]  [4]                

------------------------------------------------
logn number of divisions into two parts => n/2^k => k=logn(base2)
O(n) for merging the sorted arrays for each division
===> So time Complexity is O(n*logn) and space complextity is O(n)
"""
```

```python
def mergeSort(array, l, r):
    if l<r:
        mid = (l+r)//2
        mergeSort(array, l, mid)
        mergeSort(array, mid+1, r)
        merge(array,l,mid,r)
def merge(array, l, mid, r):
    b = []    # new array to store sorted elements
    k = 0     # for  index of new array
    i = l     # first index of subarray 1
    j = mid+1 # first index of subarray 2
    while i <= mid and j <=r:
        if array[i] <= array[j]:
            b.append(array[i])
            i += 1
        else:
            b.append(array[j])
            j += 1
        k += 1
    if i<=mid:
        b.extend(array[i:mid+1])
    else:
        b.extend(array[j:r+1])
    # copy the new array to current part of array
    array[l:r+1] = b        
```

```python
myarr = [9,4,2,13,76,34,46,12,7,0]
mergeSort(myarr, 0, 9)
print(myarr)
```
