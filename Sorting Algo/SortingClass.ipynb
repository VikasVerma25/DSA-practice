```python
class Sorting:
    def __init__(self, data):
        print("instantiated...")
        self.arr = data
        
    # Bubble Sort  - O(n^2) - best is O(n) 
    def bubble(self):        
        for i in range(len(self.arr) - 1):
            isswap = 0
            print(f"-> pass {i+1}")
            for j in range(len(self.arr)-1-i):
                if self.arr[j] > self.arr[j+1]:
                    self.arr[j],self.arr[j+1]= self.arr[j+1],self.arr[j]
                    isswap = 1
            print(self.arr)
            if isswap == 0:
                return f"Sorted: {self.arr}"
        return f"sorted: {self.arr}"
    
    #Selection Sort - O(n^2)
    def selection(self):        
        for i in range(len(self.arr) - 1):
            min = i
            print(f"-> pass {i+1}")
            for j in range(i+1, len(self.arr)):
                if self.arr[j] < self.arr[min]:                    
                    min = j
            self.arr[i],self.arr[min] = self.arr[min], self.arr[i] 
            print(self.arr)
        return self.arr
        
    #Insertion Sort - O(n^2) - best is O(n) - Good for small or nearly sorted array
    def insertion(self):
        for i in range(1,len(self.arr)-1):
            pos = i
            current = self.arr[i]
            print(f"-> pass {i}")
            while pos > 0 and self.arr[pos-1] > current:
                self.arr[pos] = self.arr[pos-1]
                pos -= 1                
            print(self.arr)
            self.arr[pos] = current            
        return f"sorted: {self.arr}"   
    
```
