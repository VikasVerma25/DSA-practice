```python
# singly linked list class - object as a linkedlist
class LinkedList:
    def __init__(self, value):
        self.head = {
            'data': value,
            'nextnode' : None
        }
        self.tail = self.head
        self.length = 1
        
    def append(self, value):
        self.tail['nextnode'] = {
            'data' : value,
            'nextnode' : None
        }
        self.tail = self.tail['nextnode']
        self.length += 1
    
    def prepend(self, value):
        newnode = {
            'data': value,
            'nextnode': self.head
        }
        self.head = newnode
        self.length += 1
        
    def insert(self, index, value):
        if not 0<=index<=self.length:
            return "Can't insert"
        elif index == 0:
            self.prepend(value)
        else:
            node = self.head
            for _ in range(index-1):
                node = node['nextnode']
            temp = node['nextnode']
            node['nextnode'] = {
                'data' : value,
                'nextnode' : temp
            }
            
            self.length += 1  
                
    def deletefirst(self):
        temp = self.head
        if temp == None:
            return "Empty"
        self.head = self.head['nextnode']
        temp['nextnode'] = None
        self.length -= 1
    
    def deletelast(self):
        temp = self.head
        if temp == None:
            return "Empty"
        elif self.length == 1:
            self.head = self.tail = None
            self.length -= 1
            return
        while True:
            if temp['nextnode'] == self.tail:
                temp['nextnode'] = None
                break
            temp = temp['nextnode']
        self.tail = temp
        self.length -= 1
    
    def delete(self, index):
        if not 0<=index<self.length:
            return "Can't delete"
        elif self.length == 1:
            self.head = self.tail = None
            self.length -= 1
        elif index == 0:
            self.deletefirst()
        elif index == self.length-1:
            self.deletelast()
        else:
            node = self.head
            for _ in range(index-1):
                node = node['nextnode']
            temp = node['nextnode']
            node['nextnode'] = temp['nextnode']
            temp['nextnode'] = None
            
            self.length -= 1 
    
    def traverse(self):
        node = self.head
        if node == None:
            return "Empty"
        while True:
            print(node['data'],' => ',end='')
            if node['nextnode'] == None:
                break
            node = node['nextnode']
            
    def reverse(self):
        self.tail = self.head
        first = self.head
        second = first['nextnode']
        while second:
            print(second['data'])
            temp = second['nextnode']
            second['nextnode'] = first
            first = second
            second = temp
            print(second)
        self.tail['nextnode'] = None
        self.head = first
            
            
```

```python
# doubly linked list class - object as a linkedlist
class dLinkedList:
    def __init__(self, value):
        self.head = {
            'data': value,
            'previous': None,
            'nextnode' : None
        }
        self.tail = self.head
        self.length = 1
    
    def append(self, value):
        self.tail['nextnode'] = {
            'data' : value,
            'previous': self.tail,
            'nextnode' : None
        }
        self.tail = self.tail['nextnode']
        self.length += 1
    
    def prepend(self, value):
        newnode = {
            'data': value,
            'previous': None,
            'nextnode': self.head
        }
        self.head['previous'] = newnode
        self.head = newnode
        self.length += 1
        
    def insert(self, index, value):
        if not 0<=index<=self.length:
            return "Can't insert"
        elif index == 0:
            self.prepend(value)
        else:
            node = self.head
            for _ in range(index-1):
                node = node['nextnode']
            temp = node['nextnode']
            node['nextnode'] = {
                'data' : value,
                'previous': node,
                'nextnode' : temp
            }
            temp['previous'] = node['nextnode']
            self.length += 1
    
    def delete(self, index):
        if not 0<=index<self.length:
            return "Can't delete"
        else:
            node = self.head
            for _ in range(index-1):
                node = node['nextnode']
            temp = node['nextnode']
            node['nextnode'] = temp['nextnode']
            node['nextnode']['previous'] = node
            temp['nextnode'] = None
            
            self.length -= 1 
    
    def traverse(self):
        node = self.head
        if node == None:
            return "Empty"
        print("Head-to-Tail")
        while True:
            print(node['data'],' => ',end='')
            if node['nextnode'] == None:
                break
            node = node['nextnode']
        print("\nTail-to-Head")
        tnode = self.tail
        while True:
            print(tnode['data'],' <= ',end='')
            if tnode['previous'] == None:
                break
            tnode = tnode['previous']
            
```
