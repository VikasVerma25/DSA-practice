```python
from math import ceil
def binary(arr, value):
    start = 0
    end = len(arr) - 1
    mid = ceil((start + end)/ 2)
    while start <= end:
        mid = ceil((start + end)/ 2)
        print(f"start:{start}, mid:{mid}, end{end}")
        if value == arr[mid]:
            return mid
        elif value > arr[mid]:
            start = mid + 1
            #binary(start, end, arr, value)
        else:
            end = mid - 1
            #binary(start, end, arr, value)
    else:
        return "Not Found"
```
