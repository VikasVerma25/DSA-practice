```python
def permutations(string, nlen=0):
    string = list(string)
    if nlen == 0:
        nlen = len(string)
    temp = string.copy()
    anslist = []
    b = ''
    for ele in range(len(string)):
        temp[len(string) - nlen] = string[ele]
        local = b.join(temp)
        anslist.append(local)
    if nlen == 1:
        return "k"
    anslist.extend(permutations(string, nlen-1))
    return anslist
```
