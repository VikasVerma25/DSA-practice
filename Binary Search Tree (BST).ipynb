```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        
def insertNode(node, root = None):   
    if root == None:
        root = node       
       
    elif node.data <= root.data:
        if root.left is None:
            root.left = node
        else:            
            insertNode(node, root.left)

    else:
        if root.right is None:
            root.right = node
        else:
            insertNode(node, root.right)
    return "Node inserted."
```

```python
a = Node(50)
```

```python
b = Node(30)
c = Node(40)
d = Node(20)
e = Node(80)
f = Node(70)
```

```python
insertNode(a)
for i in [b,c,d,e,f]:
    insertNode(i,a)
```

```python
def postOrderTraversal(mytree):
    if not mytree:
        return
    postOrderTraversal(mytree.left)
    postOrderTraversal(mytree.right)
    print(mytree.data, end=" ")
```

```python
###############################
class TreeNode:
    def __init__(self, value):
        self.data = value
        self.left = None
        self.right = None
        
class Tree:
    def __init__(self):
        self.root = None
        
    def insert(self, value):
        node = TreeNode(value)
        if not self.root:
                self.root = node
                return
        temp = self.root
        while True:            
            if node.data <= temp.data:
                if temp.left is None:
                    temp.left = node
                    break
                else:
                    temp = temp.left
            else:
                if temp.right is None:
                    temp.right = node
                    break
                else:
                    temp = temp.right
                    
    @staticmethod
    def preOrderTraversal(mytree):
        if not mytree:
            return
        print(mytree.data, end=" ")
        Tree.preOrderTraversal(mytree.left)
        Tree.preOrderTraversal(mytree.right)
```

```python
mytree = Tree()
```

```python
mytree.insert(85)
```

```python
mytree.root
```

```python
mytree.preOrderTraversal(mytree.root)
```
