```python
def permutations(lstr, l=0, anslist=[]):
    if l==0:
        anslist = []
    if l == len(lstr):
        anslist.append(''.join(lstr))
        
    for i in range(l, len(lstr)):
        lstr[l],lstr[i] = lstr[i], lstr[l]
        anslist = permutations(lstr, l+1, anslist)
        lstr[l],lstr[i] = lstr[i], lstr[l]    #backtracking
    return anslist 
    
def findpermutations(string):
    anslist = permutations(list(string))
    
    print("Permutations:")
    for i in range(len(anslist)):
        print(i+1,"-",anslist[i])
    print("Permutations (removing duplicates):")
    anslist = list(set(anslist))    
    for i in range(len(anslist)):
        print(i+1,"-",anslist[i])
```
