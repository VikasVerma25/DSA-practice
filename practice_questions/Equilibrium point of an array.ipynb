```python
def equilibriumPoint(A, N):
        if N==1:
            return 1
        prefixsum = [A[0]]
        for i in range(1,N):
            prefixsum.append(prefixsum[-1]+A[i])
        for x in range(1,N):
            if prefixsum[-1] - prefixsum[x] == prefixsum[x-1]:
                return x+1
        return -1
```
