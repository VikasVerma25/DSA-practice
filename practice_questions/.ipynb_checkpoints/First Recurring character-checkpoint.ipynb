```python
# if input: [1,3,4,5,1,9,4]
# then output: 1

# naive approach or brute force => nested forloop - O(n^2) time and O(1) space
def firstR(array):
    flag = 0
    first = len(array)
    for i in range(len(array)):        
        for j in range(i+1, len(array)):
            if array[i] == array[j]:
                flag = 1
                if j < first:
                    first = j
    if flag == 1:
        return array[first]
    return "No character is recurring"


# using hash table (set) - O(n) time and O(n) space
def firstRecurring(array):
    myset = set()
    # iterate over list and check for item inside myset
    for i in array:
        #check if character is in myset
        if i in myset:
            return i
        # add character to myset
        myset.add(i)
    
    # No recurring of character found
    else:
        return "No character is recurring"
```

```python
firstRecurring([1,4,7,2,9,0,7])
```

```python
firstR([1,4,7,2,9,0,7])
```
