```python
def permutations(lstr, l=0):
    if l == len(lstr):
        print(''.join(lstr))
        
    for i in range(l, len(lstr)):
        lstr[l],lstr[i] = lstr[i], lstr[l]
        permutations(lstr, l+1)
        lstr[l],lstr[i] = lstr[i], lstr[l]    #backtracking
        
def findpermutations(string):
    permutations(list(string))
```

```python
findpermutations('aac')
```
