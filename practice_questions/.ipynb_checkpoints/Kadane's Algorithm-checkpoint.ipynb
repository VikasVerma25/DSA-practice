maximum sum contiguous sub-array

```python
#Function to find the sum of contiguous subarray with maximum sum.
def maxSubArraySum(self,a,size):
    max_sum = 0
    max_till_here = float('-inf')
    for i in range(size):
        max_sum += a[i]
        if a[i]>max_sum:
            max_sum = a[i]
        if max_sum>max_till_here:
            max_till_here = max_sum
    return max_till_here
```
