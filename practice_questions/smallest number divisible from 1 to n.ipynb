```python
# find LCM of numbers from 1 to n
```

```python
class Solution:
    def getSmallestDivNum(self, n): 
        # code here 
        lcm = 1
        for i in range(2,n+1):
            lcm = lcm*i//Solution.gcd(lcm,i)
        return lcm  
    @staticmethod
    def gcd(a,b):
        if a==0:
            return b
        if b==0:
            return a
        if a<b:
            return Solution.gcd(a,b%a)
        else:
            return Solution.gcd(a%b,b)
```
