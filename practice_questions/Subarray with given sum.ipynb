```python
def subArraySum(arr, n, s):
        sum = 0
        j = i = 0
        while j <= i :
            f = 0
            if sum < s and i<n:
                sum += arr[i]
                i += 1
                f = 1
            elif sum > s:
                sum -= arr[j]
                j += 1
                f = 1
            if sum == s:
                return [j+1,i]
            if f == 0:
                break
        return [-1]
```
