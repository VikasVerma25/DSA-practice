Given an array arr[] of size N containing positive integers and an integer X, find the element in the array which is smaller than X and closest to it.

```python
def immediateSmaller(arr,n,x):
        diff = float('inf')
        ele = None
        for i in range(n):
            d = x-arr[i]
            if d > 0 and d < diff:
                ele = arr[i]
                diff = d
        return ele if ele else -1 
```
