```python
class Solution:
    #Function to rotate an array by d elements in counter-clockwise direction. 
    def rotateArr(self,A,D,N):
        #Your code here
        D = D%N
        self.reverse(A,0,N-1)
        self.reverse(A,N-D,N-1)
        self.reverse(A,0,N-D-1)
    def reverse(self, A, l, r):
        while(l<r):
            A[l],A[r] = A[r],A[l]
            l += 1
            r -= 1
```
