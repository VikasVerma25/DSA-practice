majority element occurs more than N/2 times in the array

```python
def findmajority(arr):
    index = 0
    count = 1
    for i in range(2, len(arr)):
        if arr[i] == arr[index]:
            count += 1
        else:
            count -= 1
        if count == 0:
            index = i
            count = 1
    # check if found element is majority or not
    if arr.count(arr[index]) > len(arr)//2:
        return arr[index]
    return "No majority element"
```

```python
print(findmajority([3,3,1,1,1,3,3,4,3,5,3]))
```
