```python
from math import ceil
def binary(array, ele, start, end):
    mid = ceil((start + end)/ 2)
    print("start:",start," mid:",mid ," end:",end)
    if start <= end:
        if array[mid] == ele:
            return(f"element found at position {mid+1}")
        elif ele > array[mid]:
            return binary(array, ele, mid+1, end)
        else:
            return binary(array, ele, start, mid-1)
    else:
        return 'not found'
```

```python
ar = [1,3,6,9,10,24,30,33,56,76,87,88,90,94,99,110]
binary(ar, 110, 0 ,len(ar)-1)
```
