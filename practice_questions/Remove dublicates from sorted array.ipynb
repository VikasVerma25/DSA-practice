```python
def remove_duplicate(A, N):
    j = 1
    for i in range(1,N):      # i and j at same position initially
        if A[i] != A[i-1]:     # if element not equal to previous then insert at j position
            A[j] = A[i]
            j += 1
    return j   # new array size (reduced or same if no dublicates)
```
