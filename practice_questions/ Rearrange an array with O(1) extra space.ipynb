```python
def arrange(arr, n): 
    #Function to rearrange an array so that arr[i] becomes arr[arr[i]]
    #with O(1) extra space.
    # mentain both old and new value at current position
    # set a[i]=(a[i]+a[a[i]]*n)
    # so, old value = a[i]%n 
    # and new value = a[i]//n
    for i in range(n):
        old = arr[arr[i]]%n
        arr[i] = arr[i]+(old*n)
    for i in range(n):
        new = arr[i]//n
        arr[i] = new
```
