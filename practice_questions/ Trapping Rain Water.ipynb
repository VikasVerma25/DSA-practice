Given an array arr[] of N non-negative integers representing the height of blocks. If width of each block is 1, compute how much water can be trapped between the blocks during the rainy season.

```python
# set pointers at both end and find the max at both end. select the minimum of both max and fill water(max-a[i]).
```

```python
def trappingWater(arr,n):
        l = 0
        l_max = arr[0]
        r = n-1
        r_max = arr[n-1]
        water = 0
        while l<r:
            if arr[l]>l_max:
                l_max = arr[l]
            if arr[r]>r_max:
                r_max = arr[r]
            if l_max<r_max:
                water += (l_max - arr[l])
                l += 1
            else:
                water += (r_max - arr[r])
                r -= 1
        return water
```
