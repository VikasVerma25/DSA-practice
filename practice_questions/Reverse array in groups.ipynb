```python
class Solution:
    #Function to reverse every sub-array group of size k.
    def reverseInGroups(self, arr, N, K):
        # code here
        l = 0
        r = K-1
        while r<N:
            self.reverse(arr,l,r)
            if r == N-1: break
            r += K
            l += K
        if r>=N:
            self.reverse(arr,l,N-1)

    def reverse(self, arr, l ,r):
        while l < r:
            arr[l],arr[r] = arr[r], arr[l];
            l += 1
            r -= 1
```
