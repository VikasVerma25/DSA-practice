```python
def merge(array1, array2):
    # if any array is empty
    if not (array1 and array2) :
        return array1+array2
    
    len1 = len(array1) 
    len2 = len(array2) 
    i = j = 0
    merge_array = []
    
    # compare the elements of both array and add to merged array
    while i<len1 and j<len2:
        if array1[i] <= array2[j]:
            merge_array.append(array1[i])
            i += 1
            print(merge_array)
        else:
            merge_array.append(array2[j])
            j += 1
            print(merge_array)
    
    # adding the remaining elements
    if i < len1:
        merge_array = merge_array + array1[i:len1]
    else:
        merge_array = merge_array + array2[j:len2]
    
    return merge_array
```

```python
merge([0,1,3,4,17,18,19,20,21,22], [1,2,3,4,5,9,12])
```
